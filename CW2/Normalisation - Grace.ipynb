```python
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, Normalizer
```

```python
# Read the dataset
all_df=pd.read_csv('TrainDataset2024.csv', index_col=False)
all_df.head() 
```

```python
# Drop the 'ID' column
all_df.drop('ID', axis=1, inplace=True)
all_df.head() 
```

```python
# Initialize scalers
scaler = StandardScaler()
normalizer = Normalizer()
```

```python
# Apply zero-mean normalization
zero_mean_data = scaler.fit_transform(all_df)

# Apply vector normalization
vector_normalized_data = normalizer.fit_transform(all_df)
```

```python
print(zero_mean_data)
```

```python
print(vector_normalized_data)
```

```python
# Displays the mean of each feature
print(scaler.mean_)
```

```python
Xs = scaler.transform(all_df)
fig,ax=plt.subplots(1,figsize=(20,8))
sns.boxplot(data=Xs,ax=ax)
```
